```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import  train_test_split
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn import metrics
import numpy as np
```

```python
olivetti_data=fetch_olivetti_faces()
```

```python
feature=olivetti_data.data
targets=olivetti_data.target
```

```python
fig, sub_plot = plt.subplots(nrows=5, ncols=8, figsize=(14,8))
sub_plot = sub_plot.flatten()
for unique_user_id in np.unique(targets):
  img_index=unique_user_id*8
  sub_plot[unique_user_id].imshow(feature[img_index].reshape(64,64), cmap='gray')
  sub_plot[unique_user_id].set_xticks([])
  sub_plot[unique_user_id].set_yticks([])
  sub_plot[unique_user_id].set_title("Face id: %s"% unique_user_id)
```

```python
fig, sub_plot = plt.subplots(nrows=40, ncols=10, figsize=(14,8))
for j in range(10):
  sub_plot[j].imshow(feature[j].reshape(64,64), cmap='gray')
  sub_plot[j].set_xticks([])
  sub_plot[j].set_yticks([])
  sub_plot[j].set_title("Face id: %s"% unique_user_id)
```

```python
X_train,X_test,y_train,y_test=train_test_split(feature,targets,test_size=0.25,stratify=targets)
```

```python
pca=PCA(n_components=100, whiten=True)
pca.fit(X_train)
```

```python
X_train_pca=pca.transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
no_of_eigenfaces=len(pca.components_)
eigen_faces=pca.components_.reshape((no_of_eigenfaces,64,64))
```

```python
fig, sub_plots = plt.subplots(nrows=10,ncols=10,figsize=(15,15))
sub_plots=sub_plots.flatten()
for i in range(no_of_eigenfaces):
  sub_plots[i].imshow(eigen_faces[i],cmap="gray")
  sub_plot[j].set_xticks([])
  sub_plot[j].set_yticks([])
```
